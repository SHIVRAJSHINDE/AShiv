```python
import requests
import pandas as pd

movie_file = pd.read_json("movies.json","r")

for movies in movie_file['data']: 
    url = "http://www.omdbapi.com/?t="+movies['name']+"&y="+str(movies['year'])+"&apikey=8a267f92"
#          http://www.omdbapi.com/?i=tt3896198&apikey=8a267f92
    r=requests.get(url)
    json_data=r.json()
    datata = {}
#to write in Json file    
    for k in json_data.keys():
        print(k+': ',json_data[k])
#to write in Json file
    for k in json_data.keys():
        datata[k]=json_data[k]
        file = open("movie_details.json","a",encoding="utf-8")
        file.write(str(datata))
        file.close()

```
