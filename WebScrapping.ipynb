```python
import matplotlib.pyplot as plt

abc = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'

df = pd.read_csv(abc,sep=';')

pd.DataFrame.hist(df.ix[:,0:1])
plt.xlabel('fixed acidity(g(tartaticc acid)/dm$^3$)')
plt.ylabel('count')
plt.show()
```

```python
# Get Titles and Text using BeautifulSoup package

import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/~guido/'

r= requests.get(url)

html_doc = r.text

soup = BeautifulSoup(html_doc,"lxml")

guido_title = soup.title

print(guido_title)

guido_text = soup.get_text()

print(guido_text)
```

```python
# Get Links from the page..
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/~guido/'

r= requests.get(url)

html_doc = r.text

soup = BeautifulSoup(html_doc,"lxml")

guido_title = soup.title

print(guido_title)

a_tags = soup.find_all('a')

for link in a_tags:
    print(link.get('href'))

```
