```python
import pandas as pd

df = pd.DataFrame({"A":{0:'a',1:'b',2:'c'},
                  'B':{0:1,1:3,2:5},
                  'C':{0:2,1:4,2:6},
                  'D':{0:7,1:9,2:11},
                  'E':{0:8,1:10,2:12}})

#CONVERT DATAFRAME USING MELT

df_melt = df.melt(id_vars=["A"],

value_vars=["B","C","D","E"],
var_name="my_vars",
value_name="my_val")

df
df_melt
```
