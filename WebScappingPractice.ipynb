```python
from urllib.request import Request, urlopen
url = "https://www.amazon.in/s?bbn=1375424031&rh=n%3A976392031%2Cn%3A%21976393031%2Cn%3A1375424031%2Cp_89%3ALenovo&dc&fst=as%3Aoff&qid=1578730398&rnid=3837712031&ref=lp_1375424031_nr_p_89_0"
abc = Request(url)
response = urlopen(abc) #Send the request and chatches the response

print("the data type of response object is: "+str((type(response))))

html = response.read() #Extract the response: html

print(html)

response.close()
```
