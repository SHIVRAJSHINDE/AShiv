```python
import matplotlib.pyplot as plt
import numpy as np
 
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,6.972]

plt.plot(year, pop)
plt.title("Year Wise Population")
plt.xlabel("Year")
plt.ylabel("Population")

tick_val = [1950,1970,1990,2010]
tick_lab = ['1950year','1970year','1990year','2010year']

plt.xticks(tick_val,tick_lab)

plt.yticks([2,3,4,5,6,7],['2B','3B','5B','6B','7B'])
np_pop = np.array(pop)
np_pop = np_pop * 20
plt.scatter(year,pop,s=np_pop)

plt.show()
```

```python
#Using xlim & ylim
#Plot the % of degrees awared to women in the compute science and physical science

import pandas as pd
df = pd.read_csv("WomenDegree.csv")
year = df["Year"]
ps = df["Physical Science"]
cs = df["Computer Science"]

plt.plot(year,cs,color='red')
plt.plot(year,ps,color='blue')

plt.xlabel('Year')
plt.ylabel('Degrees awared to women in(%)')

plt.xlim(2005,2014)
plt.ylim(10,70)

plt.title('Degrees Awared to Women(2005,2010)\nComputer Science(red)\nPhysical Science(blue)')
plt.show()
```

```python
#Using axis to control degrees xlim and ylim at a time
plt.plot(year,cs,color='blue')
plt.plot(year,ps,color='red')

plt.axis((2005,2014,10,70))
plt.show()
```

```python
#Using legend
plt.plot(year,cs,color='red',label='Computer Science')
plt.plot(year,ps,color='blue',label='Physical Science')
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('Enrollment(%)')
plt.title('Undergraduate enrollment of Women')
plt.show()
```

```python
#Using annotate()
import numpy as np
plt.plot(year,cs,color='red',label='Computer Science')
plt.plot(year,ps,color='blue',label='Physical Science')

plt.legend(loc='lower right')
csMax= cs.max()
#yrMax=np.argmax(np.array(csMax))
yrMax=year[cs.idxmax()]

plt.annotate('Maximum',xy=(yrMax,csMax),xytext=(yrMax+5,csMax+5),arrowprops=dict(facecolor='black'))

plt.xlabel('Year')
 
plt.ylabel('Enrollment(%)')
plt.title('Undergraduate enrollment of Women')
plt.show()
```

```python
#Modifying styles

plt.style.use('ggplot')
plt.subplot(2,2,1)
plt.plot(year,ps,color='blue')
plt.title('Physical Science')

plt.subplot(2,2,2)
plt.plot(year,cs,color='red')
plt.title('Computer Science')

csMax= cs.max()
yrMax=year[cs.idxmax()]

plt.annotate('Maximum',xy=(yrMax,csMax),xytext=(yrMax-1,csMax-10),arrowprops=dict(facecolor='black'))
plt.subplot(2,2,3)
plt.plot(year,ed,color='yellow')

plt.title('Education')

plt.subplot(2,2,4)
plt.plot(year,hp,color='green')

plt.title('HP')


plt.tight_layout()
plt.show()
```
